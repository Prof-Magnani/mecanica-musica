# ðŸŽ¼ AnÃ¡lise Espectral de uma Nota Musical
Este notebook demonstra a geraÃ§Ã£o e anÃ¡lise espectral de uma nota senoidal (LÃ¡ - 440 Hz), usando Python e ferramentas cientÃ­ficas. A nota Ã© representada como:

$$ x(t) = A \cdot \sin(2\pi f_0 t) $$
onde:
- $A$ Ã© a amplitude
- $f_0$ Ã© a frequÃªncia fundamental (440 Hz)
- $t$ Ã© o tempo

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# ParÃ¢metros da nota
fs = 44100  # Hz
duration = 1.0  # segundos
f0 = 440  # frequÃªncia da nota LÃ¡

# GeraÃ§Ã£o do sinal
t = np.linspace(0, duration, int(fs * duration), endpoint=False)
note = 0.5 * np.sin(2 * np.pi * f0 * t)
```

## ðŸŽ§ Forma de Onda no Tempo
Abaixo, visualizamos os primeiros 1000 pontos do sinal gerado:

```python
plt.figure(figsize=(10, 3))
plt.plot(t[:1000], note[:1000])
plt.title('Forma de onda da nota LÃ¡ (440 Hz)')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## ðŸ“ˆ Transformada de Fourier
A transformada de Fourier discreta (DFT), implementada aqui com FFT, Ã© definida por:

$$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-2\pi i kn / N} $$

```python
# CÃ¡lculo da FFT
N = len(note)
yf = fft(note)
xf = fftfreq(N, 1 / fs)

# Parte positiva do espectro
xf_pos = xf[:N // 2]
yf_pos = 2.0 / N * np.abs(yf[:N // 2])
```

## ðŸ” Espectro de FrequÃªncias
O grÃ¡fico abaixo mostra as componentes de frequÃªncia presentes na nota gerada:

```python
plt.figure(figsize=(10, 4))
plt.plot(xf_pos, yf_pos)
plt.title('Espectro da nota LÃ¡ (440 Hz)')
plt.xlabel('FrequÃªncia (Hz)')
plt.ylabel('Amplitude')
plt.xlim(0, 2000)
plt.grid(True)
plt.show()
```

## âœ… ConclusÃ£o
Observa-se um pico claro em 440 Hz, indicando a frequÃªncia fundamental da nota gerada. Esse exemplo mostra como a anÃ¡lise espectral pode revelar o conteÃºdo harmÃ´nico de um som musical.

# 🎼 Análise Espectral de uma Nota Musical
Este notebook demonstra a geração e análise espectral de uma nota senoidal (Lá - 440 Hz), usando Python e ferramentas científicas. A nota é representada como:

$$ x(t) = A \cdot \sin(2\pi f_0 t) $$
onde:
- $A$ é a amplitude
- $f_0$ é a frequência fundamental (440 Hz)
- $t$ é o tempo

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Parâmetros da nota
fs = 44100  # Hz
duration = 1.0  # segundos
f0 = 440  # frequência da nota Lá

# Geração do sinal
t = np.linspace(0, duration, int(fs * duration), endpoint=False)
note = 0.5 * np.sin(2 * np.pi * f0 * t)
```

## 🎧 Forma de Onda no Tempo
Abaixo, visualizamos os primeiros 1000 pontos do sinal gerado:

```python
plt.figure(figsize=(10, 3))
plt.plot(t[:1000], note[:1000])
plt.title('Forma de onda da nota Lá (440 Hz)')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## 📈 Transformada de Fourier
A transformada de Fourier discreta (DFT), implementada aqui com FFT, é definida por:

$$ X[k] = \sum_{n=0}^{N-1} x[n] e^{-2\pi i kn / N} $$

```python
# Cálculo da FFT
N = len(note)
yf = fft(note)
xf = fftfreq(N, 1 / fs)

# Parte positiva do espectro
xf_pos = xf[:N // 2]
yf_pos = 2.0 / N * np.abs(yf[:N // 2])
```

## 🔍 Espectro de Frequências
O gráfico abaixo mostra as componentes de frequência presentes na nota gerada:

```python
plt.figure(figsize=(10, 4))
plt.plot(xf_pos, yf_pos)
plt.title('Espectro da nota Lá (440 Hz)')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')
plt.xlim(0, 2000)
plt.grid(True)
plt.show()
```

## ✅ Conclusão
Observa-se um pico claro em 440 Hz, indicando a frequência fundamental da nota gerada. Esse exemplo mostra como a análise espectral pode revelar o conteúdo harmônico de um som musical.
